```python
movies = movies.sample(1000)
movies.head()
```

```python
import numpy as np

#MEAN OF RATING BY USER:
def block_mean(x):
    if np.max(x) - np.min(x) == 0:
        return x-x
    return (x - np.mean(x)) / (np.max(x) - np.min(x))


movies['Rating'] = movies.groupby('UserId')['Rating'].apply(block_mean)
```
