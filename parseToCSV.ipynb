```python
with open('movies.txt', 'r') as f: 
    #Get a sample of the file
    sample_text = ""
    line = ""
    for i in range(100):
        line = f.readline()
        sample_text += (line)
    while line != "\n":
        line = f.readline()
        sample_text += (line)
        
```

```python
replacements = [("product/productId: ", ""), ("review/userId: ", ","), 
              ("review/profileName: ", ","), ("review/helpfulness: ", ","), 
              ("review/score: ", ","), ("review/time: ", ","), 
              ("review/summary: ", ","), ("review/text: ", ",")]

with open('movies.txt', 'r', errors='replace') as f: 
    with open('movies.csv', 'w') as csvf:
        counter = 0
        n_reviews = 0

        for line in f:
            line = line.rstrip('\n')
            
            index = counter % 9
            if(index == 0):
                n_reviews += 1
                if n_reviews % 100000 == 0:
                    print(n_reviews)
            if(index == 8):
                csvf.write('\n')
                counter += 1
                continue
            #We will wignore the text for now
            if(index == 7 or index == 2 or index == 6):
                counter += 1
                continue
            csvf.write(line.replace(*replacements[index]))
            counter += 1
```

```python
counter = 0
n_reviews = 0
text = ""

for line in sample_text.split('\n'):
    line = line.rstrip('\n')
            
    index = counter % 9
    if(index == 8):
        text += '\n'
        counter += 1
        continue
    #We will wignore the text and name for now 
    if(index == 7 or index == 2 or index == 6):
        counter += 1
        continue
    text = text + (line.replace(*replacements[index]))
    counter += 1

print(text)    
```
